```python
import requests
import time
from tqdm import tqdm

# Initialize dictionary to store movie data
data_dictionary = {
    "adult": [],
    "backdrop_path": [],
    "genre_ids": [],
    "id": [],
    "original_language": [],
    "original_title": [],
    "overview": [],
    "popularity": [],
    "poster_path": [],
    "release_date": [],
    "title": [],
    "video": [],
    "vote_average": [],
    "vote_count": [],
    "page": []
}

# API Key (safer way to store this would be in environment variables)
API_KEY = "28dde488e69d9134fcd6eedd2bbc65c8"
MAIN_URL = "https://api.themoviedb.org/3/movie/popular"

# Loop through pages (1 to 500)
for num in tqdm(range(1, 501), desc="Fetching movie data"):
    url = f"{MAIN_URL}?api_key={API_KEY}&language=en-US&page={num}"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise error if request fails
        data = response.json()

        for movie in data.get("results", []):  # Use .get() to avoid KeyError
            data_dictionary["adult"].append(movie.get("adult", False))
            data_dictionary["backdrop_path"].append(movie.get("backdrop_path", ""))
            data_dictionary["genre_ids"].append(movie.get("genre_ids", []))
            data_dictionary["id"].append(movie.get("id", None))
            data_dictionary["original_language"].append(movie.get("original_language", ""))
            data_dictionary["original_title"].append(movie.get("original_title", ""))
            data_dictionary["overview"].append(movie.get("overview", ""))
            data_dictionary["popularity"].append(movie.get("popularity", 0))
            data_dictionary["poster_path"].append(movie.get("poster_path", ""))
            data_dictionary["release_date"].append(movie.get("release_date", ""))
            data_dictionary["title"].append(movie.get("title", ""))
            data_dictionary["video"].append(movie.get("video", False))
            data_dictionary["vote_average"].append(movie.get("vote_average", 0.0))
            data_dictionary["vote_count"].append(movie.get("vote_count", 0))
            data_dictionary["page"].append(num)

        # Sleep for a while to prevent hitting rate limits
        time.sleep(0.5)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching page {num}: {e}")
        time.sleep(2)  # Longer delay in case of error to avoid bans

print("Data collection complete!")
```
